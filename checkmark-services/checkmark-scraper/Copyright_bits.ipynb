```python
import requests
import json
```

```python
start_date = '2024-01-30'
end_date = '2024-05-24'
url = 'https://api.publicrecords.copyright.gov/search_service_external/simple_search_dsl?page_number=1&query=' + ','.join([chr(hex_value) for hex_value in range(0x61, 0x7b)]) + '&field_type=keyword&records_per_page=500&sort_order=asc&highlight=true&model='

url = url + f'&date_field=representative_date&start_date={start_date}%2000:00:00&end_date={end_date}%2000:00:00'

response = requests.get(url)

print(f'Code: {response.status_code}')
json.loads(response.text)['metadata']['hit_count']
```

Unique keys

```python
loads = []
for each in json.loads(response.text)['data'][:10]:
    test = each['hit'].copy()
    array = []

    for each_key in [x for x in test]:
        temp = each_key
        if type(test[each_key]) == dict:
            for each_key2 in test[each_key]:
                array += [temp + '_' + each_key2]
        elif type(test[each_key]) == list:
            for each_element in test[each_key]:
                if type(each_element) == dict:
                    for each_key2 in each_element:
                        array += [temp + '_' + each_key2]
                else:
                    array += [temp]
        else:
            array += [each_key]
    loads += [[x for x in array if all([x not in every for every in loads])]]
loads
```

Shared keys

```python
loads = []
for each in json.loads(response.text)['data'][:]:
    test = each['hit'].copy()
    array = []

    for each_key in [x for x in test]:
        temp = each_key
        if type(test[each_key]) == dict:
            for each_key2 in test[each_key]:
                array += [temp + '_' + each_key2]
        elif type(test[each_key]) == list:
            for each_element in test[each_key]:
                if type(each_element) == dict:
                    for each_key2 in each_element:
                        array += [temp + '_' + each_key2]
                else:
                    array += [temp]
        else:
            array += [each_key]
    if not loads:
        loads = array
    else:
        loads = [x for x in loads if x in array]
loads, len(loads)
```
